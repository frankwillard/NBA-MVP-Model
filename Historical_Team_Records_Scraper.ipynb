```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Create an empty list to store the dataframes
dfs = []

# Iterate over the years
for year in range(1946, 2023):
    # Construct the URL for each year
    url = f'https://www.landofbasketball.com/yearbyyear/{year}_{year + 1}_standings.htm'

    # Send a GET request to the website
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the table containing the standings
    if year == 1949:
      tables = soup.find_all('table')[:3]
    else:
      tables = soup.find_all('table')[:2]

    # Create standings list for the current year
    standings = []

    # Iterate over conference tables and extract record data
    standings = [{
        'year': year,
        'team': row.find_all('td')[1].text.strip(),
        'rank_in_conf': int(row.find_all('td')[0].text),
        'wins': int(row.find_all('td')[2].text),
        'losses': int(row.find_all('td')[3].text)
    } for table in tables for row in table.find_all('tr')[1:]]

    # Create a dataframe for the current year's standings
    standings_df = pd.DataFrame(standings)

    # Append the current year's standings dataframe to the list
    dfs.append(standings_df)

# Concatenate all the dataframes into a single dataframe
all_data = pd.concat(dfs, ignore_index=True)

all_data
```

```python
# Export team record data
all_data.to_csv("Historic Team Records.csv")
```
